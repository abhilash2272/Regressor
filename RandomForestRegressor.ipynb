```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
x=np.arange(1,21).reshape(-1,1)
y=np.array([
    10,12,15,18,20,25,30,28,35,40,45,42,50,55,60,58,65,70,72,75
])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor(max_depth=3,random_state=42)
dtr.fit(x_train,y_train)
from sklearn.metrics import mean_squared_error,r2_score
print(f"MSE: {mean_squared_error(y_test,dtr.predict(x_test))}")
print(f"R2: {r2_score(y_test,dtr.predict(x_test))}")
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=100,max_depth=3,random_state=42)
rfr.fit(x_train,y_train)
from sklearn.metrics import mean_squared_error,r2_score
print(f"MSE: {mean_squared_error(y_test,rfr.predict(x_test))}")
print(f"R2: {r2_score(y_test,rfr.predict(x_test))}")
```

```python
x_grid=np.linspace(x.min(),x.max(),200).reshape(-1,1)
plt.scatter(x,y,color='black')
plt.plot(x_grid,dtr.predict(x_grid),color='red')
plt.plot(x_grid,rfr.predict(x_grid),color='green')
plt.title('Decision Tree Regressionvs Random Forest Regression')
plt.xlabel('House size index')
plt.ylabel('price')
plt.legend(['Decision Tree Regression','Random Forest Regression'])
plt.show()
```
